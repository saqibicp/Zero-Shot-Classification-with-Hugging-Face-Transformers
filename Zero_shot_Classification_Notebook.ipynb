# Import libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
import transformers
from transformers import pipeline
from pylab import rcParams
```

# Set plot size:

```python
rcParams['figure.figsize'] = 10, 5
```

# Initialize the zero-shot classification pipeline:

```python
zero_shot_classifier=pipeline("zero-shot-classification")
```

# zero-shot classification:

```python
results = zero_shot_classifier(
    sequences="My favorite food is pizza and i want to travel all over the world",
    candidate_labels=["Movies","food", "travel","Animals"],
    multi_label=True
)
```

# Plot the Results

```python
plt.bar(results["labels"],results["scores"])
plt.yticks(list(np.arange(0,1,0.1)))
plt.show()
```
